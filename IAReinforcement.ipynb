```python
import math

def activation(calculo_aprendizado):
  if calculo_aprendizado >= 0:
    return 1
  return 0

bias = 1
wb = 0.5

input = 0
output_desire = 0

weight = 0.5
learning_rate = 0.01

iteracao = 0
error = math.inf

while not error ==0:
  iteracao += 1
  print("\n--> Iteração:", iteracao)

  calculo_aprendizado = (input * weight) + (bias * wb)
  print("   ", calculo_aprendizado)

  output = activation(calculo_aprendizado)
  print("\n--> Saída: ", output)

  error = output_desire - output
  if not error == 0:
    weight = weight + (learning_rate + error)
    wb = wb + learning_rate + error
    print("\n--> Peso: ", weight)
  print("__________________________________")

print("\nA máquina aprendeu")
```
